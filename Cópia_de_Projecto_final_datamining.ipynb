```python
from sklearn import datasets
import pandas as pd
```

I will import the iris dataset

```python
iris = datasets.load_iris()
```

let's get every piece of data and put it in the dataframe

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df
```

now let's generate a target, in this case, my target
because this KNN algorithm performs classification with numeric variables and not text variables

```python
df['target'] = iris.target
```

```python
df
```

let's generate a target name, just to demonstrate

```python
iris.target_names
```

in this code we are going to add a new column called "target_name" to the DataFrame df, which contains the names of the classes or species corresponding to the target values â€‹â€‹of the Iris dataset. Basically, this line of code is mapping the target values â€‹â€‹(labels) to their respective class or species names and storing these names in the new column "target_name" of the DataFrame df.

The DataFrame df now contains the characteristics of the flower instances, the corresponding target values, and the names of the classes or species associated with the target values. This will allow us to better understand the data and make more informative analysis or visualizations.

```python
df['target_name']= iris.target_names[df['target']]
```

```python
df
```

let's determine the features, that is, columns and attributes that will be part of the study.
This list is useful because you can use it to quickly access the relevant columns of the DataFrame df when performing specific operations or analysis on characteristics.

```python
iris_features = ['sepal length (cm)','sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
```

Let's determine our target(y), remembering that to make a classification using KNN we need a something number.
and it creates a y variable that stores the target values â€‹â€‹(labels) of the Iris dataset. These target values â€‹â€‹are the class or species labels corresponding to flower instances in the dataset.

```python
y=df.target
```

let's select data, the features that will be used(X)

```python
X=df[iris_features]
```

```python
X
```

```python
y
```

let's import the KNN algorithm

```python
from sklearn.neighbors import KNeighborsClassifier
```

select model
let's create a k-Nearest Neighbors (k-NN) classification model object using the KNeighborsClassifier class from the scikit-learn library. In this case, the model is configured to use 3 nearest neighbors to make classification decisions.

```python
modelo =KNeighborsClassifier(3)
```

train model
By running this line of code, the k-NN model will be fitted to the training data, i.e. it will learn how to classify instances based on features. After training, the model will be ready to make predictions using new input data using the predict function

```python
modelo.fit(X,y)
```

```python
X
```

now let's see prediction
Therefore, the output [0 0 0 ... 2 2 2 2 2 2] indicates that the model predicted the class "Setosa" (0) for the first instances, then the class "Versicolor" (1) and finally the class "Virginica" (2) for the last instances of the training set.

Remembering that using the training set to evaluate model predictions is not an ideal practice for performance evaluation, as it can result in a biased result. It is generally best to evaluate the model on a separate test suite or using cross-validation techniques.

```python
print(modelo.predict(X))
```

let's observe accuracy (percentage of correct answers), remembering that we are using the same data that were trained for validations.

The score function calculates the accuracy of the model in correctly predicting the classes of the training instances. It returns a value between 0 and 1, where 1 represents perfect precision and 0 represents very low precision.

```python
modelo.score(X,y)
```
